### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





````python
# df.loc[ '2013-01-01', 'A'] ：索引為'2013-01-01'的1列，第A欄的value(output為1個值)
# df.loc[ '2013-01-01', ['A', 'B'] ] ：索引為'2013-01-01'的1列，第A, B欄的value(output為2個值)
# df.loc[ '2013-01-01':'2013-01-02', 'A' ] ：索引為'2013-01-01', '2013-01-02'的2列，第A欄的value(output為2個值)
# df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] ：索引為'2013-01-01', '2013-01-05'的2列，第A, C欄的value(output為4個值)

#選出的資料型態也不相同。
#第一種：value
#第二種和第三種：Series
#第四種：DataFrame
````

```

# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件

import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
print('1. ')
print(df.loc[[0, 1, 2], 'col1':])
#print(df[0:3])
print('--')
print('2. ')
print(df.iloc[[1, 3, 5], 0:])
#print(df[1::2])
print('--')
print('3. ')
print(df[['col2', 'col3']])
#print(df.iloc[:, -2:])
print('--')
print('4. ')
print(df['col2'])
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df2 = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df2
````

````python
df2[df2[0] > 20]
````

````python
df2[(df2[0] + df2[1]) > 50]
````

````python
df2[(df2[0] < 30) | (df2[1] > 30)]
````

````python
df2[df2[0] + df2[1] + df2[2] + df2[3] > 100]
````
